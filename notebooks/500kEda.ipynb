# Detailed EDA: Univariate & Bivariate Analysis
Because the features `f0-f11` are mathematically stripped of meaning, our goal is not to interpret what they *are* in English, but rather how they are *distributed* mathematically.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make the graphs look professional
plt.style.use('ggplot')

# 1. Load the perfectly balanced 500k Sample
data = pd.read_csv(r"../data/sampled_uplift.csv")
print(f"Data Loaded: {data.shape[0]} rows ready for analysis")
```

## 1. Univariate Analysis (Single Feature Distributions)
We plot histograms of `f0-f11`.

**What to look for:** Are they normally distributed (a perfect bell curve), or are they extremely skewed? If a feature is highly skewed, we might have to use a `StandardScaler` in Phase 2 so XGBoost doesn't get confused by massive outliers.

```python
# Plot histograms for the 12 anonymous features
features = [f'f{i}' for i in range(12)]

data[features].hist(bins=50, figsize=(15, 10), color='royalblue', edgecolor='black')
plt.suptitle("Univariate Analysis: Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

## 2. Bivariate Analysis (Feature vs Target Correlations)
We use a Heatmap to see how the mathematical features directly relate to the `conversion` target.

**What to look for:** Do any of the features `f0-f11` strongly correlate with `conversion` (or each other)? If a feature has a highly positive or highly negative number connected to `conversion`, it means it's incredibly powerful signal for our AI!

```python
# Plot a correlation heatmap
plt.figure(figsize=(12, 8))

# Calculate mathematical correlation between features, treatment, and our conversion target
correlation_matrix = data[features + ['treatment', 'conversion']].corr()

sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Bivariate Analysis: Feature Correlations", fontsize=16)
plt.show()
```
