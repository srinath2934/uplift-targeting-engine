# Step 1: Data Understanding & Visualization

In this notebook, our only goal is to look at the math visually using Histograms and Heatmaps. We will NOT clean or change the data here.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_theme(style="whitegrid")

# Load the pristine 500k Sample
data = pd.read_csv(r"../data/sampled_uplift.csv")
print(f"Data Loaded: {data.shape[0]} rows ready for visual analysis")

feature_cols = [f'f{i}' for i in range(12)]
```

### A. Univariate Histograms (Checking for Skewness)
Look at the shape of each feature. Notice how some are heavily squished to the left? This tells us we **must** scale our data in the next notebook.

```python
data[feature_cols].hist(bins=50, figsize=(15, 10), color='royalblue', edgecolor='black')
plt.suptitle("Feature Distributions (Notice the extreme outliers?)", fontsize=16)
plt.tight_layout()
plt.show()
```

### B. Bivariate Correlation Heatmap (Feature vs Conversion)
Let's see the mathematical correlations. If a square is dark blue or dark red, that feature is extremely important to the AI!

```python
plt.figure(figsize=(12, 8))
correlation_matrix = data[feature_cols + ['treatment', 'conversion']].corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Mathematical Correlations with Conversion", fontsize=16)
plt.show()
```
